```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'])
df['Release_Date'] = df['Release_Date'].dt.year
```

```python
cols=['Overview','Original_Language','Poster_Url']
df.drop(cols,axis=1,inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
def popularity(df,cols,labels):

    edges=[df[cols].describe()['min'],
          df[cols].describe()['25%'],
          df[cols].describe()['50%'],
          df[cols].describe()['75%'],
          df[cols].describe()['max']]
           
    df[cols]=pd.cut(df[cols],edges,labels=labels)
    return df

labels=['not_popular','below_avg','average','polpular']

popularity(df,'Vote_Average',labels)
```

# One GENRE per row

```python
df['Genre']= df['Genre'].str.split(', ')

df=df.explode('Genre').reset_index(drop=True)
df.head()
```

```python
df['Genre']=df['Genre'].astype('category')
```

# DATA VISUALIZATION


# 1) What is the most frequent Genre of movie released on netflix?

```python
df['Genre'].value_counts()
```

```python
genre_counts = df['Genre'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Count']
genre_counts
```

```python
plt.style.use('ggplot')
```

```python
sns.barplot(y='Genre', x='Count', data=genre_counts,
            order = df['Genre'].value_counts().index,
            color='lightcoral',
           edgecolor='Black')
plt.title('Genre column Distribution')
plt.tight_layout()
```

# 2) Which has highest votes in vote average column?

```python
votes=df['Vote_Average'].value_counts().reset_index()
votes.columns=['Vote_Average','Votes']
votes
```

```python
sns.barplot(y='Vote_Average' ,x='Votes', data=votes,
           order=df['Vote_Average'].value_counts().index,
           color='skyblue',
           edgecolor='Black')

plt.xlabel('VOTES')
plt.ylabel('VOTE AVERAGE')
plt.title('Votes Distribution')
```
